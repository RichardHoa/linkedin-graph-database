```python
# Clean the raw_data.csv to remove duplicate entries
import polars as pl

# 1. Load the raw data
input_file = "./Data/original_data.csv"
output_file = "./Data/nam_processed_data.csv"

df = pl.read_csv(input_file)

df_cleaned = df.filter(pl.col("id").is_unique())

df_cleaned.write_csv(output_file)

removed_count = df.height - df_cleaned.height
print(f"Original row count: {df.height}")
print(f"Cleaned row count:  {df_cleaned.height}")
print(f"Total rows removed: {removed_count}")
print(f"File saved as:      {output_file}")
```

```python
# Inspect a row in nam-data
import polars as pl
import json
import sys

def process_csv_polars(file_path):
    try:
        # 1. Efficient loading with Polars (multithreaded)
        df = pl.read_csv(file_path)

        # 2. Print all headers
        print("### HEADERS ###")
        for col in df.columns:
            print(f"• {col}")
        print("\n" + "="*40 + "\n")

        # 3. Find the row with the most non-null values
        # Calculate horizontal sum of non-null values per row
        df_with_counts = df.with_columns(
            non_null_count=pl.sum_horizontal(pl.all().is_not_null())
        )
        
        # Sort and get the top row
        most_populated_row = df_with_counts.sort("non_null_count", descending=True).limit(1)
        
        # Convert to dictionary for easy iteration
        row_dict = most_populated_row.to_dicts()[0]
        # Remove the helper count column before printing
        row_dict.pop("non_null_count", None)

        # 4. Pretty print with JSON detection
        print(f"### ROW WITH MAXIMUM DATA ###\n")
        
        for key, value in row_dict.items():
            print(f"{key}:")
            
            if isinstance(value, str):
                trimmed = value.strip()
                if (trimmed.startswith('{') and trimmed.endswith('}')) or \
                   (trimmed.startswith('[') and trimmed.endswith(']')):
                    try:
                        print(json.dumps(json.loads(trimmed), indent=4))
                    except (ValueError, TypeError):
                        print(value if value is not None else "NULL")
                else:
                    print(value if value is not None else "NULL")
            else:
                print(value if value is not None else "NULL")
            
            print("\n----\n")

    except Exception as e:
        print(f"Error processing file: {e}")

process_csv_polars("./Data/nam_processed_data.csv")
```
