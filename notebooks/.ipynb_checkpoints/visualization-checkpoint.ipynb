```python
import polars as pl
import plotly.graph_objects as go

pl.Config.set_tbl_rows(1000)
pl.Config.set_tbl_width_chars(1000)

# 1. Load and aggregate transitions
lf = pl.scan_csv("../Data/final.csv")
df_counts = (
    lf.select(["first_industry", "current_industry"])
    .drop_nulls()
    .group_by(["first_industry", "current_industry"])
    .len()
    .sort("len", descending=True)
    .head(40)  # Focus on top 40 flows for clarity
    .collect()
)

# 2. Create unique labels for nodes
# We append suffix to distinguish 'First' from 'Current' if the name is the same
sources = [f"{x} (Start)" for x in df_counts["first_industry"]]
targets = [f"{x} (Current)" for x in df_counts["current_industry"]]
values = df_counts["len"].to_list()

all_nodes = list(dict.fromkeys(sources + targets))
node_indices = {name: i for i, name in enumerate(all_nodes)}

# 3. Map names to indices
source_indices = [node_indices[s] for s in sources]
target_indices = [node_indices[t] for t in targets]

# 4. Generate Sankey Diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=all_nodes,
        color="royalblue"
    ),
    link=dict(
        source=source_indices,
        target=target_indices,
        value=values,
        color="rgba(100, 149, 237, 0.4)" # Transparent cornflower blue
    )
)])

fig.update_layout(title_text="Career Flow: First Industry to Current Industry", font_size=10)
fig.write_html("industry_sankey.html")
print("Sankey diagram saved as industry_sankey.html")
```

```python
import polars as pl
import json

# 1. Initialize LazyFrame for memory efficiency
lf = pl.scan_csv("../Data/final.csv")

# 2. Extract and process the 'educations' column
# We use a custom function to parse the JSON strings if they are complex
def extract_universities():
    # Collect the column into memory (educations is a mandatory field)
    df = lf.select("educations").collect()
    
    all_universities = []
    
    for row in df["educations"]:
        try:
            # Parse the JSON string into a Python list
            edu_entries = json.loads(row)
            
            # Ensure it is a list of education objects
            if isinstance(edu_entries, list):
                for edu in edu_entries:
                    name = edu.get("university")
                    if name:
                        all_universities.append(name.strip())
            elif isinstance(edu_entries, dict):
                name = edu_entries.get("university")
                if name:
                    all_universities.append(name.strip())
        except (json.JSONDecodeError, TypeError):
            continue

    # 3. Create a Polars Series and count frequencies
    u_series = pl.Series("university", all_universities)
    top_10 = (
        u_series.value_counts()
        .sort("count", descending=True)
        .head(10)
    )
    
    return top_10

if __name__ == "__main__":
    result = extract_universities()
    print("Top 10 Universities by Enrollment:")
    print(result)
```

```python
import polars as pl
import json

pl.Config.set_tbl_rows(1000)
pl.Config.set_fmt_str_lengths(1000)
pl.Config.set_tbl_width_chars(1000)

# Define path to your dataset
DATA_PATH = "../Data/final.csv"

def get_top_degrees():
    # 1. Scan CSV and collect only the 'educations' column
    # Polars scan_csv is optimized for memory usage on large files (510k+ rows)
    lf = pl.scan_csv(DATA_PATH)
    raw_df = lf.select("educations").collect()
    
    extracted_degrees = []
    
    # 2. Iteratively parse JSON strings
    for row in raw_df["educations"]:
        try:
            data = json.loads(row)
            
            # Ensure the data is iterable (list of education objects)
            if isinstance(data, list):
                for entry in data:
                    deg = entry.get("major_standardized")
                    if deg:
                        extracted_degrees.append(deg.strip().lower())
            elif isinstance(data, dict):
                deg = data.get("major_standardized")
                if deg:
                    extracted_degrees.append(deg.strip().lower())
        except (json.JSONDecodeError, TypeError):
            continue

    # 3. Aggregate results using Polars Series
    top_20 = (
        pl.Series("major_standardized", extracted_degrees)
        .value_counts()
        .sort("count", descending=True)
        .head(100)
    )
    
    return top_20

if __name__ == "__main__":
    result = get_top_degrees()
    print("Top 20 Standardized Major:")
    print(result)
```

```python
import polars as pl
import json

# --- CONFIGURATION TO PREVENT TRUNCATION ---
pl.Config.set_tbl_rows(1000)              # Show all 20 rows
pl.Config.set_fmt_str_lengths(1000)     # Show full industry names
pl.Config.set_tbl_width_chars(2000)     # Prevent table wrapping
# ------------------------------------------

FILE_PATH = "../Data/final.csv"

def get_top_standardized_industries():
    # 1. Load only the 'jobs' column for efficiency
    # Total records to process: 510,743
    lf = pl.scan_csv(FILE_PATH)
    df = lf.select("jobs").collect()

    industry_list = []

    # 2. Extract values from nested JSON
    # industry_standardized is mandatory within the jobs schema
    for row in df["jobs"]:
        try:
            job_data = json.loads(row)
            
            # Jobs column is a list of job objects
            if isinstance(job_data, list):
                for job in job_data:
                    ind = job.get("industry_standardized")
                    if ind:
                        industry_list.append(ind.strip())
            elif isinstance(job_data, dict):
                ind = job_data.get("industry_standardized")
                if ind:
                    industry_list.append(ind.strip())
        except (json.JSONDecodeError, TypeError):
            continue

    # 3. Aggregate and sort
    top_20 = (
        pl.Series("Industry Standardized", industry_list)
        .value_counts()
        .sort("count", descending=True)
    )
    
    return top_20

if __name__ == "__main__":
    result = get_top_standardized_industries()
    print(result)
```

```python
import polars as pl
import json

# Define the dataset path
FILE_PATH = "../Data/final.csv"

# Change this variable to your target industry
TARGET_INDUSTRY = "libraries, archives, museums and other cultural activities" 

def get_specific_industry_stats(target_name):
    # 1. Load the 'jobs' column
    lf = pl.scan_csv(FILE_PATH)
    df = lf.select("jobs").collect()

    user_inputs = set()
    job_count = 0
    found = False

    # 2. Iterate through records
    for row in df["jobs"]:
        try:
            job_list = json.loads(row)
            if isinstance(job_list, list):
                for job in job_list:
                    std = job.get("industry_standardized", "").strip().lower()
                    
                    if std == target_name.lower():
                        found = True
                        job_count += 1
                        # Capture the user input for this specific match
                        usr = job.get("industry_user_input")
                        if usr:
                            user_inputs.add(usr.strip().lower())
        except (json.JSONDecodeError, TypeError):
            continue

    # 3. Output as structured bullet points
    if not found:
        print(f"No records found for industry: '{target_name}'")
        return

    print(f"\nREPORT FOR: {target_name.upper()}")
    print("-" * 40)
    print(f"* Total Jobs in Dataset: {job_count}")
    print(f"* User-Input Variations ({len(user_inputs)} unique):")
    for variation in sorted(user_inputs):
        print(f"  - {variation}")

if __name__ == "__main__":
    get_specific_industry_stats(TARGET_INDUSTRY)
```

```python
import polars as pl
import json

# Configuration
SEARCH_TERM = "museum operations"

def find_individuals_by_industry(target_input):
    # 1. Scan only 'id' and 'jobs' columns to optimize performance
    lf = pl.scan_csv(FILE_PATH)
    df = lf.select(["id", "jobs"]).collect()

    matched_ids = []

    # 2. Parse nested JSON job entries
    for row in df.iter_rows(named=True):
        try:
            job_list = json.loads(row["jobs"])
            if isinstance(job_list, list):
                for job in job_list:
                    # Check the industry_user_input field
                    usr_input = job.get("industry_user_input", "").strip().lower()
                    
                    if usr_input == target_input.lower():
                        matched_ids.append(row["id"])
                        break  # Found a match for this person, move to next individual
        except (json.JSONDecodeError, TypeError):
            continue

    # 3. Output results
    if not matched_ids:
        print(f"No individuals found with user input: '{target_input}'")
    else:
        print(f"Found {len(matched_ids)} individual(s):")
        for profile_id in matched_ids:
            print(f"- {profile_id}")

if __name__ == "__main__":
    find_individuals_by_industry(SEARCH_TERM)
```
