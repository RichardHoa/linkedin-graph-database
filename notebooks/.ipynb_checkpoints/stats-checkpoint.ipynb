```python
import polars as pl
import json

def lookup_person(file_path, target_id, id_column="id"):
    try:
        # Filter lazily and collect only the relevant row
        df = pl.scan_csv(file_path).filter(pl.col(id_column) == target_id).collect()
        
        if df.height > 0:
            print(f"--- Records Found: {df.height} match(es) for ID '{target_id}' in {file_path} ---")
            
            records = df.to_dicts()
            
            for i, record in enumerate(records, 1):
                print(f"Result {i}:")
                
                # Pre-process the record to parse any JSON strings found in fields
                processed_record = {}
                for key, value in record.items():
                    if isinstance(value, str):
                        # Clean up common Polars/CSV artifacts before parsing
                        stripped_val = value.strip()
                        if (stripped_val.startswith('[') and stripped_val.endswith(']')) or \
                           (stripped_val.startswith('{') and stripped_val.endswith('}')):
                            try:
                                processed_record[key] = json.loads(value)
                            except json.JSONDecodeError:
                                processed_record[key] = value
                        else:
                            processed_record[key] = value
                    else:
                        processed_record[key] = value

                # Print the fully nested structure with indentation
                print(json.dumps(processed_record, indent=4, ensure_ascii=False))
                print("." * 20)
                
            print("-" * 30)
        else:
            print(f"Status: No matches found for ID '{target_id}' in {file_path}.")
            
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# --- Configuration & Execution ---
ID_TO_FIND = "melissa-jardine-phd-b8230b21"
COLUMN_NAME = "id"
files = ["../Data/nam_processed_data.csv", "../Data/final.csv"]

for file in files:
    lookup_person(file, ID_TO_FIND, COLUMN_NAME)
    print("\n" + "="*40 + "\n")
```

```python
import polars as pl
import json
import os

def generate_llm_data_profile_stream(file_path, batch_size=100_000):
    print(f"Starting profile generation for: {file_path}")
    
    # 1. Preliminary Scan: Get total rows and schema efficiently
    try:
        lf = pl.scan_csv(file_path)
        total_rows = lf.select(pl.len()).collect().item()
    except Exception as e:
        print(f"Error reading file metadata: {e}")
        return

    # Initialize profile structure
    profile = {
        "dataset_metadata": {
            "filename": file_path,
            "total_records": total_rows
        },
        "columns": {}
    }

    # Tracking structures for incremental aggregation
    col_trackers = {}

    # 2. Batched Processing
    reader = pl.read_csv_batched(file_path, batch_size=batch_size)
    
    batch_idx = 0
    while True:
        batches = reader.next_batches(1) 
        if not batches:
            break
        
        df_batch = batches[0]
        batch_idx += 1
        
        for col in df_batch.columns:
            if col not in col_trackers:
                col_trackers[col] = {
                    "total_non_null": 0,
                    "is_json": False,
                    "checked_type": False,
                    "json_key_counts": {}, 
                    "json_total_entries": 0,
                    "max_key_len": -1,
                    "best_example": None,
                    "scalar_example": None,
                    "unique_values": set(),          # Outer unique values
                    "nested_unique_values": {}       # Inner JSON unique values
                }
            
            tracker = col_trackers[col]
            series = df_batch[col].drop_nulls()
            
            if series.is_empty():
                continue

            tracker["total_non_null"] += len(series)

            # Determine column type using the first non-null value encountered
            if not tracker["checked_type"]:
                first_val = str(series[0]).strip()
                tracker["scalar_example"] = first_val
                if first_val.startswith(("{", "[")):
                    tracker["is_json"] = True
                tracker["checked_type"] = True

            # Process Outer Unique Values for non-JSON columns
            if not tracker["is_json"]:
                tracker["unique_values"].update(series.cast(pl.Utf8).to_list())

            # If it's a JSON column, process row-by-row
            if tracker["is_json"]:
                for val in series.to_list():
                    try:
                        data = json.loads(val)
                        entries = data if isinstance(data, list) else [data]
                        
                        for entry in entries:
                            if not isinstance(entry, dict): continue
                            
                            tracker["json_total_entries"] += 1
                            current_keys = [k for k, v in entry.items() if v is not None and str(v).strip() != ""]
                            
                            for k, v in entry.items():
                                # Update global key counts
                                increment = 1 if k in current_keys else 0
                                tracker["json_key_counts"][k] = tracker["json_key_counts"].get(k, 0) + increment

                                # Track unique values for the nested key
                                if k not in tracker["nested_unique_values"]:
                                    tracker["nested_unique_values"][k] = set()
                                
                                if v is not None:
                                    tracker["nested_unique_values"][k].add(str(v))

                            # Track best example (most populated fields)
                            if len(current_keys) > tracker["max_key_len"]:
                                tracker["max_key_len"] = len(current_keys)
                                tracker["best_example"] = entry
                    except:
                        continue

    # 3. Finalize Aggregation
    print("Finalizing profile...")
    
    for col, tracker in col_trackers.items():
        presence_pct = (tracker["total_non_null"] / total_rows) * 100 if total_rows > 0 else 0
        
        if tracker["total_non_null"] == 0:
             profile["columns"][col] = {"status": "all_null", "presence_pct": 0}
             continue

        if tracker["is_json"]:
            field_analysis = {}
            total_json_entries = tracker["json_total_entries"]
            
            if total_json_entries > 0:
                for key, count in tracker["json_key_counts"].items():
                    nested_vals = list(tracker["nested_unique_values"].get(key, []))
                    nested_unique_count = len(nested_vals)
                    
                    field_info = {
                        "presence_pct": round((count / total_json_entries) * 100, 5),
                        "is_mandatory": (count / total_json_entries) == 1.0,
                        "unique_count": nested_unique_count
                    }
                    
                    if nested_unique_count < 100:
                        field_info["unique_values_list"] = nested_vals
                        
                    field_analysis[key] = field_info
            
            profile["columns"][col] = {
                "type": "json_structure", 
                "column_presence_pct": round(presence_pct, 2),
                "nested_schema": field_analysis,
                "representative_example": tracker["best_example"]
            }
        else:
            unique_vals = list(tracker["unique_values"])
            unique_count = len(unique_vals)
            
            col_data = {
                "type": "scalar_text",
                "column_presence_pct": round(presence_pct, 2),
                "unique_count": unique_count,
                "example_value": tracker["scalar_example"]
            }
            
            if unique_count < 100:
                col_data["unique_values_list"] = unique_vals
                
            profile["columns"][col] = col_data

    # Output to File
    output_dir = "../Data"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        
    output_path = os.path.join(output_dir, "final_structure.json")
    with open(output_path, "w") as f:
        json.dump(profile, f, indent=2)
    
    print(f"Profile saved to: {output_path}")

# Run
if __name__ == "__main__":
    generate_llm_data_profile_stream("../Data/final.csv")
```

```python
import polars as pl

# Load the datasets
ha_path = "../Data/ha_processed_data.csv"
nam_path = "../Data/nam_processed_data.csv"

# Read CSVs and select unique IDs immediately
# Replace 'id' with your actual column name if different
df_ha_unique = pl.read_csv(ha_path).select(pl.col("id").unique())
df_nam_unique = pl.read_csv(nam_path).select(pl.col("id").unique())

# Use Anti-Join to find IDs in HA not present in NAM
ha_not_in_nam = df_ha_unique.join(df_nam_unique, on="id", how="anti")

# Use Anti-Join to find IDs in NAM not present in HA
nam_not_in_ha = df_nam_unique.join(df_ha_unique, on="id", how="anti")

# Output Results
print(f"Summary Statistics:")
print(f"{'-'*20}")
print(f"Total Unique IDs in HA:  {df_ha_unique.height}")
print(f"Total Unique IDs in NAM: {df_nam_unique.height}")
print(f"IDs in HA but not in NAM: {ha_not_in_nam.height}")
print(f"IDs in NAM but not in HA: {nam_not_in_ha.height}")

# Print 100 example IDs for each case
print("\nExample 100 IDs (HA but not NAM):")
if not ha_not_in_nam.is_empty():
    print(ha_not_in_nam.head(30)["id"].to_list())
else:
    print("None found.")

print("\nExample 100 IDs (NAM but not HA):")
if not nam_not_in_ha.is_empty():
    print(nam_not_in_ha.head(30)["id"].to_list())
else:
    print("None found.")
```

```python
import polars as pl

# 1. Load the dataset
# Adjust the path if your script is in a different directory
df = pl.read_csv("../Data/final.csv")

# 2. Define the schema for the certifications JSON string
# This matches the structure you provided in final_structure.json
cert_schema = pl.List(pl.Struct([
    pl.Field("title", pl.String),
    pl.Field("subtitle", pl.String),
    pl.Field("meta", pl.String),
    pl.Field("credential_id", pl.String),
    pl.Field("credential_url", pl.String)
]))

# 3. Filter for profiles that have certifications and sample 10 people
random_10_people = (
    df.filter(pl.col("certifications").is_not_null())
    .sample(10)
)

import polars as pl

# 1. Load the dataset using the path from your structure file
df = pl.read_csv("../Data/final.csv")

# 2. Define the nested schema based on your final_structure.json
cert_schema = pl.List(pl.Struct([
    pl.Field("title", pl.String),
    pl.Field("subtitle", pl.String),
    pl.Field("meta", pl.String),
    pl.Field("credential_id", pl.String),
    pl.Field("credential_url", pl.String)
]))

# 3. Filter for profiles with certifications and grab 10 random people
# We parse the JSON string into a structured object immediately
sampled_df = (
    df.filter(pl.col("certifications").is_not_null() & (pl.col("certifications") != "[]"))
    .sample(10)
    .with_columns(
        pl.col("certifications").str.json_decode(cert_schema)
    )
    .select(["id", "certifications"])
)

# 4. Iterate and print in the requested bulleted format
for row in sampled_df.to_dicts():
    person_id = row["id"]
    certs = row["certifications"]
    
    print(f"Person ID: {person_id}")
    
    if certs:
        for i, cert in enumerate(certs, 1):
            print(f"  Certification {i}:")
            print(f"  • Title: {cert.get('title')}")
            print(f"  • Subtitle: {cert.get('subtitle')}")
            print(f"  • Meta: {cert.get('meta')}")
            print(f"  • Credential ID: {cert.get('credential_id')}")
            print(f"  • Credential URL: {cert.get('credential_url')}")
    else:
        print("  • No certifications found.")
        
    print("-" * 50)
```
